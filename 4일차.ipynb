- x1, x2 : 입력 값
- w1, w2 : 가중치 = 신호를 얼마나 증폭할 지, x1과 x2에 대하여 곱해줌
- w0 : bias = 입력값과 상관없지 무조건 들어오는 값
- y : 출력 값

-  y = activation(w0 + w1x1 + w2x2) : 활성화 함수
   -  x >= 0 일 때 1
   -  x < 0 일 때 0

#### 퍼셉트로 예시


```python
 def Perceptron(x_1,x_2):
    
    # 설정한 가중치값을 적용
    w_0 = -5 
    w_1 = -1
    w_2 = 5
    
    # 활성화 함수에 들어갈 값을 계산
    output = w_0+w_1*x_1+w_2*x_2
    
    # 활성화 함수 결과를 계산
    if output < 0:
        y = 0
    else:
        y = 1
    
    return y, output
```

#### 다층 퍼셉트론
- 비선형적인 문제 해결 
- 단층 퍼셉트론을 여러 층으로 쌓은 것
   - 히든층 : 입력층과 출력층 사이의 Layer
   - 히든층이 많아지면 깊은 신경망이라는 의미 DL

- 히든층 
   - 장점 : 분류가 정확하고 많이 할 수 있다.
   - 단점 : 구해야하는 가중치가 많아진다.

#### 다층 퍼셉트론 구현하기

```python
def perceptron_layer1(X):
    w1 = 0.2
    w2 = 0.5
    w3 = -0.4
    w4 = 0.9
    w5 = 0.8
    w6 = 0
    Y = []
    Y.append(X[0] * w1 + X[1] * w2 + X[2] * w3)
    Y.append(X[0] * w4 + X[1] * w5 + X[2] * w6)
    return Y
    
def perceptron_layer2(X):
    w7 = 0.2
    w8 = 0.8
    w9 = 0.9
    w10 = -0.3
    Y = []
    Y.append(X[0] * w7 + X[1] * w8)
    Y.append(X[0] * w9 + X[1] * w10)
    return Y
    
def perceptron_layer3(X):
    w11 = 0.1
    w12 = 0.3
    return X[0]*w11 + X[1]*w12
    
X = [0.4, 0.2, 0.3] # 임의의 데이터
Y1 = perceptron_layer1(X)
Y2 = perceptron_layer2(Y1)
Y = perceptron_layer3(Y2)

print(Y)
```
